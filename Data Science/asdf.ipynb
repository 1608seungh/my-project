# 전처리 과정 - 알파벳을 숫자로, object를 정수형 실수형으로

```python
train.loc[train['target'] =='o', 'target'] = 1 ## 알파벳을 숫자로 바꾸는 과정  
train.loc[train['target'] =='x', 'target'] = 0
train
```

```python
trainx = train.iloc[:, :-1]
trainy = train['target']
```

```python
trainy
```

```python
trainy = trainy.astype('int64').to_numpy()
trainy
```

```python
from sklearn.model_selection import train_test_split ## o,x가 순서대로 되어 있어서 섞어주기 위한 과정 
x_train, x_test, y_train, y_test = train_test_split(trainx, trainy, test_size = 0.3, random_state = 12)
x_train.shape, x_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x_train, y_train) 
```

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score

def get_metrics(y_test, y_predicted):
    display(pd.crosstab(pd.Series(y_test, name='Actual'), pd.Series(y_predicted, name='Predicted')).\
            style.background_gradient(cmap='Reds'))
    
    accuracy = accuracy_score(y_test, y_predicted)
    precision = precision_score(y_test, y_predicted, average='weighted')
    recall = recall_score(y_test, y_predicted, average='weighted')
    f1 = f1_score(y_test, y_predicted, average='weighted')
    
    print(f'\naccuracy \t= {round(accuracy, 2)} \
    \nprecision \t= {round(precision, 2)} \
    \nrecall \t\t= {round(recall, 2)} \
    \nf1-score \t= {round(f1, 2)}')
```

```python
get_metrics(y_test, model.predict(x_test)) ## 성능 평가
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train, y_train) 
```

```python
get_metrics(y_test, model.predict(x_test)) ## 성능 평가
```

```python
data = pd.read_csv("C:/Users/A/Downloads/a.csv")
data
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
x = data['P'].values.reshape(-1,1) ## 데이터를 모델에 넣기 위해서는 펼쳐줘야 함
```

```python
knn = NearestNeighbors(n_neighbors = 3)
knn.fit(x)
```

```python
distance, index = knn.kneighbors(x)
```

```python
index
```

```python
plt.plot(distance.mean(axis = 1));
```

```python
outlier_index = np.where(distance.mean(axis = 1) >= 1)
```

```python
outlier_index
```

```python
outlier = data.iloc[outlier_index]
outlier
```
